```python
import requests
import sqlite3

# Connect to SQLite database (or create it if it doesn't exist)
conn = sqlite3.connect('FinalDBMaltese.db')
cursor = conn.cursor()

# Create table if it doesn't exist
cursor.execute('''
CREATE TABLE IF NOT EXISTS library_cdn_links (
    id INTEGER PRIMARY KEY,
    library_name TEXT NOT NULL,
    version TEXT NOT NULL,
    cdn_link TEXT NOT NULL
)
''')
conn.commit()

libraries = [
    'jquery',
    'dojo',
    'd3',
    'pixi.js',
    'three.js',
    'velocity-animate',
    'react',
    'vue',
    'ember',
    'angular.js'
]

def check_library_existence(library):
    """ Check if a library exists on CDNJS """
    response = requests.get(f"https://api.cdnjs.com/libraries/{library}")
    if response.status_code == 200:
        return True  # Library exists
    return False  # Library does not exist or other error

def fetch_cdn_info(library):
    """ Fetch CDN info if library exists """
    try:
        response = requests.get(f"https://api.cdnjs.com/libraries/{library}")
        if response.status_code == 200:
            data = response.json()
            assets = data['assets']
            cdn_links = []
            for asset in assets:
                version = asset['version']
                files = asset['files']
                for file in files:
                    cdn_link = f"https://cdnjs.cloudflare.com/ajax/libs/{library}/{version}/{file}"
                    cdn_links.append((version, file, cdn_link))
                    #cursor.execute('INSERT INTO library_cdn_links (library_name, version, cdn_link) VALUES (?, ?, ?)', (library, version, cdn_link))
            conn.commit()
            return cdn_links
        else:
            return []
    except Exception as e:
        print(f"An error occurred: {str(e)}")
        return []

for lib in libraries:
    if check_library_existence(lib):
        cdn_links = fetch_cdn_info(lib)
        if cdn_links:
            for version, file, cdn_link in cdn_links:
                print(f"{lib}: Version {version}, File {file}, CDN Link {cdn_link}")
        else:
            print(f"No versions or files found for {lib}")
    else:
        print(f"Library not found: {lib}")

# Close the database connection
conn.close()
```

<h1>Downloading all JS Lib Versions</h1>

```python
import requests
import os
import time
from concurrent.futures import ThreadPoolExecutor

# Define libraries to fetch
libraries = [
    'jquery',
    'dojo',
    'd3',
    'pixi.js',
    'three',
    'velocity-animate',
    'react',
    'vue',
    'ember',
    'angular',
]
failedDownloads =[]
downloaded = []
# Function to fetch library versions
def fetch_versions(library):
    response = requests.get(f"https://data.jsdelivr.com/v1/package/npm/{library}")
    if response.status_code == 200:
        versions = response.json().get('versions', [])
        return versions
    else:
        print(f"Failed to fetch versions for {library}")
        return []

# Function to download a specific version of a library
def download_library(library_version):
    library, version = library_version
    lib_dir = f'libraries/{library}/{version}'
    file_path = os.path.join(lib_dir, f"{library}-{version}.js")
    
    # Check if the file already exists
    if os.path.exists(file_path):
        print(f"{library}@{version} is already downloaded.")
        downloaded.append(f"{library}@{version}")
        return
    
    # Proceed with download if the file does not exist
    url = f"https://cdn.jsdelivr.net/npm/{library}@{version}"
    response = requests.get(url)
    if response.status_code == 200:
        # Create the directory for the library if it doesn't exist
        os.makedirs(lib_dir, exist_ok=True)
        # Write the response content to a file
        with open(file_path, 'wb') as file:
            file.write(response.content)
        print(f"Downloaded {library}@{version}")
        downloaded.append(f"{library}@{version}")
    else:
        print(f"Failed to download {library}@{version}")
        failedDownloads.append(f"{library}@{version}")

# Function to manage downloads for each library
def manage_downloads(library):
    print(f"Fetching versions for {library}...")
    versions = fetch_versions(library)
    with ThreadPoolExecutor(max_workers=5) as executor:
        # Create a list of tuples for the library and its versions
        library_versions = [(library, version) for version in versions]
        # Schedule the download tasks and execute them in parallel
        executor.map(download_library, library_versions)

# Main function to start the download process
def main():
    with ThreadPoolExecutor(max_workers=5) as executor:
        # Schedule the manage_downloads tasks for each library
        executor.map(manage_downloads, libraries)
    
    print("Failed Downloads")
    for fd in failedDownloads:
        print(fd)
    #for d in downloaded:
    #    print(d)
    print(f"Failed Downloads{len(failedDownloads)}")
    print(f"Downloaded{len(downloaded)}")
    print(f"Total{len(failedDownloads)+len(downloaded)}")    

if __name__ == "__main__":
    main()
```
